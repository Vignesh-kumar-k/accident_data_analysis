```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import pickle
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
accident_df=pd.read_csv('Accidents0514.csv')
```

```python
accident_df.columns
```

```python
q3_df=pd.DataFrame(data=accident_df,columns=['Longitude','Latitude','Junction_Detail','Location_Easting_OSGR','Location_Northing_OSGR','Urban_or_Rural_Area','Pedestrian_Crossing-Human_Control'])
```

```python
q3_df
```

```python
X=q3_df.drop('Pedestrian_Crossing-Human_Control',axis=1)
y=q3_df['Pedestrian_Crossing-Human_Control']
q3_df['Pedestrian_Crossing-Human_Control'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree= DecisionTreeClassifier()
```

```python
dtree.fit(X_train.values,y_train.values)
```

```python
predictions= dtree.predict(X_test.values)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test.values,predictions))
```
