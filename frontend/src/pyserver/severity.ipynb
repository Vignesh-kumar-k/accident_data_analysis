```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

df=pd.read_csv('Accidents0514.csv')
```

```python
df.isnull().values.any()
```

```python
fatal_df=pd.DataFrame(data=df,columns=['Speed_limit','Day_of_Week','Local_Authority_(District)','Weather_Conditions','Accident_Severity'])
```

```python
X=fatal_df[['Weather_Conditions','Local_Authority_(District)']]
y=fatal_df['Accident_Severity']
fatal_df['Accident_Severity'].unique()
fatal_df.tail(20)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y)

```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree= DecisionTreeClassifier()
```

```python
dtree.fit(X_train.values,y_train.values)
```

```python
predictions= dtree.predict(X_test.values)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test.values,predictions))
```

```python
import pickle
pickle.dump(dtree, open("severity.pkl", "wb"))
```
